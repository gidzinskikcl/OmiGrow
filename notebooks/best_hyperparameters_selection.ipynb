```python
import pandas as pd
```

```python
filenames = [
    "tuning_results_0_40.csv",
    "tuning_results_40_80.csv",
    "tuning_results_80_120.csv",
    "tuning_results_120_160.csv",
    "tuning_results_160_200.csv",
    "tuning_results_200_240.csv",
    "tuning_results_240_280.csv",
    "tuning_results_280_320.csv",
]
expr_dir = "../results/predictions/results_S1_ID_40_expr/"

expr_df = pd.DataFrame()

for f in filenames:
    df = pd.read_csv(expr_dir + f)
    expr_df = pd.concat([expr_df, df], ignore_index=True)

expr_df
```

```python
# MAE ranking
mae_df = expr_df.sort_values(by="mean_MAE", ascending=True)
mae_df.head(5)
```

```python
# RMSE ranking
rmse_df = expr_df.sort_values(by="mean_RMSE", ascending=True)
rmse_df.head(5)
```

```python
# MDAE ranking
mdae_df = expr_df.sort_values(by="mean_MDAE", ascending=True)
mdae_df.head(5)
```

```python
# PCC ranking
pcc_df = expr_df.sort_values(by="mean_PCC", ascending=False)
pcc_df.head(5)
```
